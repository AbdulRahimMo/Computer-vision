```python
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def plt_imshow(title, image):
  # convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
def mse(imageA, imageB):
	 #defining mse function, which takes two arguments: imageA and imageB
  #convert the images from unsigned 8-bit integers to floating point
  #dividing the sum of squares by the total number of pixels in the image
	err = np.sum((imageA.astype("float") - imageB.astype("float"))** 2)
	err /= float(imageA.shape[0] * imageA.shape[1])
	
	# return the MSE, the lower the error, the more "similar"
	# the two images are
	return err

def compare_images(imageA, imageB, title):
	# compute the mean squared error and structural similarity
	# index for the images
	m = mse(imageA, imageB)
	s = ssim(imageA, imageB)

	# setup the figure
	fig = plt.figure(title)
	plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))

	# show first image
	ax = fig.add_subplot(1, 2, 1)
	plt.imshow(imageA, cmap = plt.cm.gray)
	plt.axis("off")

	# show the second image
	ax = fig.add_subplot(1, 2, 2)
	plt.imshow(imageB, cmap = plt.cm.gray)
	plt.axis("off")
 
	# show the images
	plt.show()
```

```python
# load the images -- the original, the original + contrast,
# and the original + photoshop
BackHand = cv2.imread("/content/BackHand.jpg")
BackHand_2 = cv2.imread("/content/BackHand_2.jpg")
BackHand_3 = cv2.imread("/content/BackHand_3.jpg")
BackHand_4 = cv2.imread("/content/BackHand_4.jpg")
BackHand_5 = cv2.imread("/content/BackHand_5.jpg")
BackHand_6 = cv2.imread("/content/BackHand_6.jpg")
BackHand_7 = cv2.imread("/content/BackHand_7.jpg")
```

```python
#Resizing the pictures
BackHand=cv2.resize(BackHand,(1000,1000))
BackHand_2=cv2.resize(BackHand_2,(1000,1000))
BackHand_3=cv2.resize(BackHand_3,(1000,1000))
BackHand_4=cv2.resize(BackHand_4,(1000,1000))
BackHand_5=cv2.resize(BackHand_5,(1000,1000))
BackHand_6=cv2.resize(BackHand_6,(1000,1000))
BackHand_7=cv2.resize(BackHand_7,(1000,1000))
```

```python
# convert the images to grayscale
BackHand = cv2.cvtColor(BackHand, cv2.COLOR_BGR2GRAY)
BackHand_2 = cv2.cvtColor(BackHand_2, cv2.COLOR_BGR2GRAY)
BackHand_3 = cv2.cvtColor(BackHand_3, cv2.COLOR_BGR2GRAY)
BackHand_4 = cv2.cvtColor(BackHand_4, cv2.COLOR_BGR2GRAY)
BackHand_5 = cv2.cvtColor(BackHand_5, cv2.COLOR_BGR2GRAY)
BackHand_6 = cv2.cvtColor(BackHand_6, cv2.COLOR_BGR2GRAY)
BackHand_7 = cv2.cvtColor(BackHand_7, cv2.COLOR_BGR2GRAY)
```

```python
# initialize the figure
fig = plt.figure("Images")
images = ("Back1", BackHand), ("Back2", BackHand_2), ("Back3", BackHand_3), ("Back4 with Ring", BackHand_4), ("Back5 with Ring", BackHand_5),("Back6 with Ring",BackHand_6),("Back7 with Ring",BackHand_7)

# loop over the images
for (i, (name, image)) in enumerate(images):
	# show the image
	ax = fig.add_subplot(1, 7, i + 1)
	#ax.set_title(name)
	plt.imshow(image, cmap = plt.cm.gray)
	plt.axis("off")

# show the figure
plt.show()
```

```python
# compare the same images
compare_images(BackHand, BackHand, "Back1 vs.Back1")
compare_images(BackHand_2, BackHand_2, "Back2 vs. Back2")
compare_images(BackHand_5, BackHand_5, "Back5 with Ring vs. Back5 with Ring")
```

```python
# compare the different images
compare_images(BackHand, BackHand_3, "Back1 vs. Back3 ")
compare_images(BackHand_4, BackHand_2, "Back4 with Ring vs. Back2")
compare_images(BackHand_5, BackHand_4, "Back5 with Ring vs. Back4 with Ring")
```
